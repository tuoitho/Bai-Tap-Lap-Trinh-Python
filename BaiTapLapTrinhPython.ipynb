```python
def khoang_cach_chinh_sua(str1, str2, m, n):
    if m == 0:
        return n
    if n == 0:
        return m

    if str1[m-1] == str2[n-1]:
        return khoang_cach_chinh_sua(str1, str2, m-1, n-1)

    return 1 + min(khoang_cach_chinh_sua(str1, str2, m, n-1),        # Thêm
                   khoang_cach_chinh_sua(str1, str2, m-1, n),        # Xóa
                   khoang_cach_chinh_sua(str1, str2, m-1, n-1))      # Thay thế

# Hàm chính
if __name__ == "__main__":
    chuoi1 = input("Nhập chuỗi thứ nhất: ")
    chuoi2 = input("Nhập chuỗi thứ hai: ")

    kq = khoang_cach_chinh_sua(chuoi1, chuoi2, len(chuoi1), len(chuoi2))
    print(f"Khoảng cách chỉnh sửa giữa hai chuỗi {chuoi1} và {chuoi2} là:", kq)
```

```python
def find_most_frequent_character(input_string):
    # Create a dictionary to store the frequency of each character
    char_freq = {}
    
    # Count the frequency of each character in the string
    for char in input_string:
        if char in char_freq:
            char_freq[char] += 1
        else:
            char_freq[char] = 1
    
    # Find the most frequently occurring character and its frequency
    most_frequent_char = max(char_freq, key=char_freq.get)
    frequency = char_freq[most_frequent_char]
    # return the most frequently occurring character and its frequency like the output in the description of the problem
    return most_frequent_char, frequency
if __name__ == "__main__":
    input_string = input("Enter a string: ")
    kq = find_most_frequent_character(input_string)
    print(kq)

```

```python
def calculate_parity(data):
    if len(data) != 8:
        print("Lỗi: Chuỗi không đúng độ dài (phải có 8 bit)")
        return None
    # Đếm số bit 1 trong chuỗi
    count_ones = sum(int(bit) for bit in data)
    # Tính toán bit chẵn lẻ
    parity_bit = 0 if count_ones % 2 == 0 else 1
    return parity_bit

def main():
    while True:
        data_input = input("Nhập vào một nhóm 8 bit (nhập dòng trống để kết thúc): ")
        # Kiểm tra điều kiện kết thúc
        if data_input == "":
            break
        # Kiểm tra độ dài của chuỗi
        if len(data_input) != 8:
            print("Lỗi: Chuỗi không đúng độ dài (phải có 8 bit)")
            continue
        # Kiểm tra các bit nhập vào có phải là 0 hoặc 1 không
        if not all(bit.isdigit() and int(bit) in [0, 1] for bit in data_input):
            print("Lỗi: Chuỗi chỉ chứa các bit 0 hoặc 1")
            continue
        # Tính toán bit chẵn lẻ
        parity_bit = calculate_parity(data_input)
        if parity_bit is not None:
            print(f"Bit chẵn lẻ phải là: {parity_bit}")

if __name__ == "__main__":
    main()
```
